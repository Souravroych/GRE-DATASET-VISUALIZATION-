```python
cols=df.drop(labels='Serial No.',axis=1)
cols.head()

df_corr = df.corr()
plt.figure(figsize=(12,10))
ax = sns.heatmap(df_corr, annot=True, cmap='viridis') #notation: "annot" not "annote"
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.title('Correlation Matrix', fontsize=16);

# plt.figure(figsize=(12,10))
# sns.heatmap(cols.corr(),cmap='viridis',annot=True,linewidths=2.0)
# plt.show();
```

The above shows the correlation matrix between the columns and as we can see the highest correlation of chance of admission is with GRE,TOEFL SCORE & LOR
